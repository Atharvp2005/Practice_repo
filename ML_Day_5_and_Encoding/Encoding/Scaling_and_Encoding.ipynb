```python
x = df2_for_scaling[['CLMAGE' , 'LOSS']]
x
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = StandardScaler() # transforms the value near the 0 the values can be +ve , -ve , greater than 1 , or greater thsn -1 , etc.it is normal.
# it has mean = 0 and standard deviation = 1.
x_std = sc.fit_transform(x)
```

```python
x_std
```

```python
minMax_scale = MinMaxScaler() # transforms the values between 0 and 1.
x_minMax = minMax_scale.fit_transform(x)
```
