```python
import pandas as pd
import numpy as np
```

```python
data = [23,5,23,42,4]
s1 = pd.Series(data)
print(f"The series is :\n{s1}")
print(s1[3])

# Series syntax is Series(array , indexing)
print("The result of the student is : ")
studentName = ["Ram" , "Xyz" , "Hello" , "Balram"] # This is the custom index created 
marks = [99,98,89,70] 
result = pd.Series(marks , studentName)
print(result)
print(f"The marks of \"Ram\" are : {result["Ram"]}") # accessing the series using custom index created
```

```python
# create a series using nd array 
numArray = np.array([23,6,52,6,5224])
s2 = pd.Series(numArray , index=[1,2,3,4,5])
print(s2)
```

## DataFrame

```python
df1 = pd.DataFrame([23,6,4,6,5,4])
print(df1)
```

```python
df2 = pd.DataFrame([[23,65,24,24,4] , [45,34,5,43,55]])
print(df2)
```

```python
# using columns method to see the total columns
df3 = pd.DataFrame([{"Name":"Person1" , "Age":34} , {"Name":"Person2" , "Age":56}])
print(df3)
print(f"The Name of the columns are : {df3.columns}")
```

```python
# creating a empty dataframe
dataFrame1 = pd.DataFrame()
print(dataFrame1)
```

```python
dataFrame2 = pd.DataFrame([25,32,24,4,2])
print(dataFrame2)
```

```python
# Creating Dataframe by giving column names
demoData = [['stud1' , 10] , ['stud2' , 20] , ['stud3' , 30]]
df4 = pd.DataFrame(demoData , columns = ['Name' , 'Age'])
print(df4)
```
