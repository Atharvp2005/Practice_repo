```python
x = df[['sex' , 'cp' , 'thalach' , 'exang' , 'oldpeak' , 'slope' , 'ca' , 'thal']]
y = df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x , y , train_size = 0.8 , random_state = 123456)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion = 'gini' , max_depth = 500)
model.fit(x_train , y_train)
```

```python
from sklearn.model_selection import GridSearchCV # the grid search model is used to find the best parameters for the model

paprameters = {
    'criterion' : ['gini' , 'entropy' , 'log-loss'] , 
    "max_depth" : np.arange(10) * 100
}

grid_search_cv = GridSearchCV(estimator=model , param_grid=paprameters)

```

```python
grid_search_cv.fit(x_train , y_train)
```

```python
print(f"The best parameter is : {grid_search_cv.best_params_}") # it is used to find the best hyper parameter for the model.
```

```python
print(f"The best score is : {grid_search_cv.best_score_}")
```

```python
y_pred = grid_search_cv.predict(x_test) # prediction using new grid_search_cv / decision tree model .
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score , precision_score , recall_score , f1_score , roc_auc_score

print(f"The confusion matrix is : \n{confusion_matrix(y_test , y_pred)}")
print(f"The accuracy score is : {accuracy_score(y_test , y_pred)*100:.2f}%")
print(f"The precision score is : {precision_score(y_test , y_pred)*100:.2f}%")
print(f"The recall score is : {recall_score(y_test , y_pred)*100:.2f}%")
print(f"The f1 score is : {f1_score(y_test , y_pred)*100:.2f}%")
print(f"The roc auc score is : {roc_auc_score(y_test , y_pred)*100:.2f}%")
```

```python
plt.scatter(x_test['thalach'][y_pred == 0] , x_test['oldpeak'][y_pred == 0] , label = 'Won\'t suffer' , color = "green")
plt.scatter(x_test['thalach'][y_pred == 1] , x_test['oldpeak'][y_pred == 1] , label = "Will suffer" , color = "red")
plt.xlabel("Thalach")
plt.ylabel("Old Peak")
plt.title("Thalach vs Old Peak")
plt.legend()
```
