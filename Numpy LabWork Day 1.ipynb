## Numpy Assignment

```python
import numpy as np
```

```python
a1 = np.array([2,4,2,5])
print(a1)
print(a1.itemsize)
print(a1.dtype)
```

```python
a2 = np.array([23,5,8,4,3] , dtype = np.int32)
print(a2.itemsize)
print(a2.dtype)
```

```python
a3 = np.array([236,56,32,6,23] , dtype = np.int16)
print(a3.itemsize)
print(a3.dtype)
```

```python
a4 = np.array([2,56,32,6,23] , dtype = np.int8)
print(a4.itemsize)
print(a4.dtype)
```

```python
a5 = np.array([23,6,3,5,6])
print(f"The size of array is : {a5.size}")
```

```python
a6 = np.array([23,5,3,4,5])
print(f"The total space occupied in memory is : {a6.nbytes}")
```

```python
a7 = np.array([2,54,4,3])
print(a7.ndim)

a8 = np.array([[2,54,4,3] , [2,5,5,3]])
print(a8.ndim)
```
